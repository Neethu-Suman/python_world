###II. Scatter Plots
  
  Generate scatter plots with:

    plt.scatter(x_values, y_values)

```python
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
plt.scatter(x, y, color='red', marker='o')
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# make the data
x = 4 + np.random.normal(0, 2, 24)
y = 4 + np.random.normal(0, 2, len(x))
# size and color:
sizes = np.random.uniform(15, 80, len(x))
colors = np.random.uniform(15, 80, len(x))

# plot
fig, ax = plt.subplots()

ax.scatter(x, y, s=sizes, c=colors, vmin=0, vmax=100)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

###III. Bar Plots

###V. Pie Charts
  
  Create pie charts with:

    plt.pie(values, labels)

```python
plt.pie([10, 20, 30, 40], labels=['A', 'B', 'C', 'D'])
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for the pie chart
labels = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [10, 30, 20, 40]
colors = ['lightblue', 'lightcoral', 'lightgreen', 'lightpink']

# Create a pie chart
plt.figure(figsize=(6,6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
#Auto-label slices - Pass a function or format string to autopct to label slices.

# Add a title
plt.title('Distribution of Categories')

# Add a legend
plt.legend(labels, loc='upper right')

# Display the pie chart
plt.show()
```

##5. Subplots

### Creating Multiple Plots
  Use plt.subplot() to create multiple plots in the same figure:
    
    plt.subplot(rows, cols, index)

###Advanced Subplots

For more advanced subplot layouts, use:

    fig, axes = plt.subplots(rows, cols)

```python
plt.subplot(2, 1, 1)
plt.plot([1, 2, 3, 4], [10, 20, 25, 30])

plt.subplot(2, 1, 2)
plt.scatter([1, 2, 3, 4], [10, 20, 25, 30])

plt.show()
```

```python
import numpy as np

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.subplot(2, 1, 1)
plt.plot(x, y1)
plt.title('Subplot 1: Sin(x)')

plt.subplot(2, 1, 2)
plt.plot(x, y2)
plt.title('Subplot 2: Cos(x)')

plt.tight_layout()
plt.show()
```

##6. Customizing Plots

###Colors, Markers, and Linestyles

  * Customize plot aesthetics with various options:

        plt.plot(x, y, color='blue', marker='o', linestyle='--')

  * Adding Legends
    
    Include legends for better plot interpretation:

        plt.legend(['Label 1', 'Label 2'])

  * Setting Axis Limits and Ticks

      Control the axis limits and ticks for better presentation:

        plt.xlim(min_val, max_val)
        
        plt.xticks(tick_values)

```python
# Customizing plots
plt.plot([1, 2, 3, 4], [10, 20, 25, 30], color='green', linestyle='dashed', marker='o', markersize=8)

# Adding legend
plt.legend(['Data'])

# Setting axis limits and ticks
plt.xlim(0, 5)
plt.ylim(0, 35)
plt.xticks([1, 2, 3, 4])
plt.yticks([0, 10, 20, 30])

plt.show()
```

```python
"""
Customizing plots is crucial for conveying information effectively. Let's explore customization options.
"""

# Example: Customizing Plot
import numpy as np

x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label='Sin(x)', color='blue', linestyle='--', marker='o')
plt.plot(x, y2, label='Cos(x)', color='green', linestyle='-', marker='s')

plt.title('Customized Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.xlim(0, 2*np.pi)
plt.ylim(-1, 1)
plt.xticks(np.arange(0, 2*np.pi+1, np.pi/2), ['0', '/2', '', '3/2', '2'])
plt.grid(True)
plt.show()
```

##7. Annotations and Text

  Annotations and text help in providing additional context to the plots. Let's see how to use them.

```python
# Adding text and annotations
plt.plot([1, 2, 3, 4], [10, 20, 25, 30])
plt.text(2, 22, 'Important Point', fontsize=12, color='red')
plt.annotate('Max Value', xy=(4, 30), xytext=(3.5, 27), arrowprops=dict(facecolor='black', shrink=0.05))
plt.show()
```

```python
# Example: Annotations and Text
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)
plt.plot(x, y, label='Sin(x)', color='blue')

# Annotate a specific point
plt.annotate('Peak', xy=(np.pi/2, 1), xytext=(np.pi/2, 0.8),arrowprops=dict(facecolor='black', shrink=0.05))

# Add text
plt.text(np.pi, 0, 'Zero Crossing', ha='center', va='bottom')

plt.title('Annotations and Text')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()
```

##8. Saving and Exporting Plots

  Saving figures using

    plt.savefig()

  Choosing the appropriate file format (PNG, PDF, SVG, etc.).

```python
# Saving figures
plt.plot([1, 2, 3, 4], [10, 20, 25, 30])
plt.savefig('my_plot.png', dpi=300)

# Choosing the appropriate file format
# Supported formats: 'png', 'pdf', 'svg', etc.
plt.savefig('my_plot.pdf')

plt.show()
```

```python
"""
Once you've created a plot, it's essential to save it for further use or sharing. Let's explore saving options.
"""

# Example: Saving and Exporting Plots
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

plt.plot(x, y, label='Sin(x)', color='blue')
plt.title('Save and Export Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)

# Save the plot as a PNG file
plt.savefig('sin_plot.png')
# Alternatively, save as PDF or SVG for vector graphics

plt.show()
```
